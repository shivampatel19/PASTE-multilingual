```python
import json
```

```python
# Opening JSON file
f = open('dev.json')
```

```python
# returns JSON object as
# a dictionary
data = json.load(f)
```

```python
data
```

```python
data[1]['text']
```

```python
# Iterating through the json
# list
file = open('dev.sent', 'w')
for i in data:
    print(i['text'])
    file.write(i['text']+'\n')
```

```python
# Closing file
f.close()
file.close()
```
